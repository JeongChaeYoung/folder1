```python
from ultralytics import YOLO
import os
import pandas as pd
import matplotlib.pyplot as plt

print(" [êµì°¨ ì¼ë°˜í™”] ì „ë°© â†” ìˆ˜ì§ ê²€ì¦ ì‹œì‘")

RESULT_ROOT = r"D:/aihub_down/060.ê³ í•´ìƒë„ ë„ë¡œë…¸ë©´ ì´ë¯¸ì§€ ë°ì´í„°/3.ê°œë°©ë°ì´í„°/1.ë°ì´í„°/YOLO_dataset/results/yolov11"

DATA_FRONTAL = r"D:/aihub_down/060.ê³ í•´ìƒë„ ë„ë¡œë…¸ë©´ ì´ë¯¸ì§€ ë°ì´í„°/3.ê°œë°©ë°ì´í„°/1.ë°ì´í„°/YOLO_dataset/data_frontal.yaml"
DATA_VERTICAL = r"D:/aihub_down/060.ê³ í•´ìƒë„ ë„ë¡œë…¸ë©´ ì´ë¯¸ì§€ ë°ì´í„°/3.ê°œë°©ë°ì´í„°/1.ë°ì´í„°/YOLO_dataset/data_vertical.yaml"

# í•™ìŠµëœ best.pt (ì´ë¯¸ ì €ì¥ëœ ëª¨ë¸)
best_frontal = os.path.join(RESULT_ROOT, "frontal_train", "weights", "best.pt")
best_vertical = os.path.join(RESULT_ROOT, "vertical_train", "weights", "best.pt")

# êµì°¨ ê²°ê³¼ ì €ì¥ í´ë”
CROSS_SAVE_ROOT = os.path.join(RESULT_ROOT, "cross_results")
os.makedirs(CROSS_SAVE_ROOT, exist_ok=True)

# ê³µí†µ YOLO ì„¤ì •
IMGSZ = 640
DEVICE = 0  # GPU

# 1) ëª¨ë¸ ë¡œë“œ
print("best frontal:", best_frontal)
print("best vertical:", best_vertical)

model_frontal = YOLO(best_frontal)   # ì „ë°© í•™ìŠµëª¨ë¸
model_vertical = YOLO(best_vertical) # ìˆ˜ì§ í•™ìŠµëª¨ë¸

# 2) êµì°¨ ê²€ì¦ ì‹¤í–‰

print("ì „ë°© ëª¨ë¸ â†’ ìˆ˜ì§ ë°ì´í„° ê²€ì¦")

results_frontal_on_vertical = model_frontal.val(
    data=DATA_VERTICAL,
    imgsz=IMGSZ,
    device=DEVICE,
    project=CROSS_SAVE_ROOT,
    name="frontal_on_vertical"
)

print("ìˆ˜ì§ ëª¨ë¸ â†’ ì „ë°© ë°ì´í„° ê²€ì¦")

results_vertical_on_frontal = model_vertical.val(
    data=DATA_FRONTAL,
    imgsz=IMGSZ,
    device=DEVICE,
    project=CROSS_SAVE_ROOT,
    name="vertical_on_frontal"
)

print("êµì°¨ ì¼ë°˜í™” ê²€ì¦ ì™„ë£Œ")


# 3) êµì°¨ mAP, Precision, Recall ìš”ì•½ ì •ë¦¬

cross_fv = results_frontal_on_vertical.box   # Frontal train â†’ Vertical val
cross_vf = results_vertical_on_frontal.box   # Vertical train â†’ Frontal val

summary = {
    "Trainâ†’Val":   ["Frontalâ†’Vertical", "Verticalâ†’Frontal"],
    "mAP50":       [cross_fv.map50,      cross_vf.map50],
    "mAP50-95":    [cross_fv.map,        cross_vf.map],
    "Precision":   [cross_fv.mp,         cross_vf.mp],
    "Recall":      [cross_fv.mr,         cross_vf.mr],
}

df = pd.DataFrame(summary)

csv_path = os.path.join(CROSS_SAVE_ROOT, "cross_generalization_summary.csv")
df.to_csv(csv_path, index=False, encoding="utf-8-sig")

print("êµì°¨ ì¼ë°˜í™” ì„±ëŠ¥ CSV ì €ì¥:", csv_path)
print(df)

# 4) ì„±ëŠ¥ ë¹„êµ ê·¸ë˜í”„ ì €ì¥
plt.figure(figsize=(8, 5))
plt.bar(df["Trainâ†’Val"], df["mAP50"], label="mAP50")
plt.bar(df["Trainâ†’Val"], df["mAP50-95"], alpha=0.7, label="mAP50-95")
plt.title("YOLOv11 êµì°¨ ì¼ë°˜í™” ì„±ëŠ¥ ë¹„êµ (Frontal â†” Vertical)")
plt.ylabel("Score")
plt.ylim(0, 1.0)
plt.grid(alpha=0.3)
plt.legend()

graph_path = os.path.join(CROSS_SAVE_ROOT, "cross_generalization_plot.png")
plt.savefig(graph_path, dpi=300, bbox_inches="tight")
plt.show()

print("ê·¸ë˜í”„ ì €ì¥ ì™„ë£Œ:", graph_path)
```
